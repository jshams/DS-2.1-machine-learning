## Clean the data

1. Get rid of white space in column names
2. Drop unnecessary columns
3. numerate non-numeric columns

```python
# Get rid of spaces in column names
def remove_spaces(word):
    return ''.join([letter for letter in word if letter != ' '])
def rename_cols_w_no_spaces(df):
    new_cols = {}
    for column in df.columns:
        new_cols[column] = remove_spaces(column)
    return df.rename(columns=new_cols)
df = rename_cols_w_no_spaces(df)
df.columns
```
